## rjust, ljust, center

- these three are used for the alignment of the text
- rjust() ---- it gives the space before the string
- ljust() --- it gives the space after the string
- center() --- it gives the equal space before and after the string

```python
"hello".rjust(10)    # it gives the space before the string
```

```python
"hello".ljust(10)        # it gives the space after the string
```

```python
"hello".center(10)         # it gives the equal space before and after the string
```
