- 1. compile time error
- 2. Logical type error
- 3. Run time error

- try --- when the try block gets error then it execute the except block 
- when it not gives error it skips the exception block

```python
c = 3
d = 0

try:
    print(c/d)

except Exception:
    print('hey you cannot divide number by zero')

print('bye')    
```

```python
c = 3
d = 0

try:
    print(c/d)

except Exception as e:
    print('hey you cannot divide number by zero:',e)

print('bye')    
```

```python
c = 3
d = 0

try:
    print(d/c)

except Exception:
    print('hey you cannot divide number by zero')

print('bye')    
```

- in real time when you open the database you must need to close the database

```python
c = 3
d = 3

try:
    print('resource open')
    print(c/d)
    print('resource closed')

except Exception as e:
    print('hey you cannot divide number by zero:',e)
```
