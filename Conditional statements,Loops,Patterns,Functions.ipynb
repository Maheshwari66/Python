# GLOBALS()

- Globals() is a built-in fn that returns a dictionary represnting the current global system table.it allows you to access and modify global variables programmatically

```python
x = 10     # global variable

def update_x():
    global x          # global variable considered as local 
    x += 5                # modify the global variable

update_x()
print(x)
```

```python
x = 10     # global variable

def update_x():
    #global x          
    x += 5                # modify the global variable

update_x()
print(x)                 # it gives error bcoz we are not mention the local           
```

```python
x = 10           # global variable

def update_x():
    globals()['x'] += 5    # access and modify the global variable using the dictionary

update_x()
print(x)
```

# Pass list to function

- can you pass list of elements in the function that function will return the count of even or odd number from the list

```python
def count(lst):
    
    even = 0
    odd = 0
    
    for i in lst:
        if i % 2 == 0:
            even += 1
        else:
            odd += 1
    return even ,odd

lst = [10,9,8,23,50]
even,odd = count(lst)

print(even)
print(odd)
```

# Factorial 

```python
def fact(n):           #0 never comes in factorial
    f = 1
    for i in range(1,n+1):
        f = f*i

    return f

x = 4
result = fact(x)
print(result)                # 4 * 3 * 2 * 1
```

# Factorial number using recursion

```python
def fact(n):
    if n==0:
        return 1
    return n * fact(n-1)

result = fact(5)
result
```
