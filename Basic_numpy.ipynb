## Reshaping the methods in numpy

#### 1. transpose

- converting from rows to columns and columns to rows

```python
P = np.arange(12).reshape(3,4)
P
```

```python
P.transpose()
```

```python
P.T
```

#### 2 . Ravel 

- convert it any dimensions to one dimension

```python
P.ravel()
```

#### 3 . Stacking 

- joining the arrays of same dimensions
- stackings are 2 types
- 1 . vstack = vertical stack
- 2 . hstack = horizontal stack

```python
P
```

```python
P1 = np.arange(12,24).reshape(3,4)
P1
```

```python
np.hstack((P,P1))                         # horizontal stack
```

```python
np.vstack((P,P1))                      # vertical stack
```

#### 4 . Splitting

- opposite to the stacking 
- splitting are two types
- 1 . vsplit
- 2 . hsplit

```python
P
```

```python
np.hsplit(P,2)
```

```python
P1
```

```python
np.vsplit(P1,3)
```
