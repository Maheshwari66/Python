# Swap Variables

- in python there are 3 ways to swap the variables
- 1 . by using the third variable (temp)
- 2 . ( a , b = b , a )
- 3 . ( a = a + b 
        b = a - b 
        a = a - b )

```python
aa = 5
bb = 6
```

```python
aa = bb
bb = aa
```

```python
print (aa)
print (bb)
```

```python
# in the above we lost the value 5
aa1 = 7
bb1 = 8
```

```python
temp = aa1                       # we swap the variables by using the 3rd variable
aa1 = bb1
bb1 = temp 
```

```python
print(aa1)
print(bb1)
```

```python
aa2 = 9
bb2 = 3                   # second way
```

```python
aa2 , bb2 = bb2 , aa2
```

```python
print(aa2)
print(bb2)
```

```python
aa3 = 45
bb3 = 56
```

```python
aa3 = aa3 + bb3
bb3 = aa3 - bb3                   # third way to swap the variable
aa3 = aa3 - bb3
```

```python
print(aa3)
print(bb3)
```
