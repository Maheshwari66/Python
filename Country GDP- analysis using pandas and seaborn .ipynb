## Plotting the graphs using matplotlib and seaborn

```python
import matplotlib.pyplot as plt      # matplotlib is used for visualization
import seaborn as sns               # seaborn is used for statistical visualization

%matplotlib inline
plt.rcParams['figure.figsize']=6,2

import warnings 
warnings.filterwarnings('ignore')         # os error
```

```python
df['InternetUsers']
```

```python
vis1 = sns.distplot(df["InternetUsers"])   #univarient analysis- plot the graph using one variable
plt.show()
```

```python
vis2= sns.displot(df["InternetUsers"] )        # displot it shows histogram
plt.show()
```

```python
vis3 = sns.distplot(df["InternetUsers"],bins = 15)
plt.show()                       #bins-used for captures the clear visualization
```

```python
#Box plot 

vis4 = sns.boxplot(data = df, x = 'IncomeGroup' , y = 'BirthRate')     
plt.show()                                                       # bivarient analysis
```

```python
vis5 = sns.lmplot(data = df, x = 'InternetUsers' , y = 'BirthRate')    #bivarient analysis
plt.show()
```

```python
vis5=sns.lmplot(data=df,x='InternetUsers',y ='BirthRate',fit_reg=False)   
plt.show()                                          

# fig_reg = false it doesnt shows the line,
# bydefault the system shows line without writing the true
```

```python
vis6=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False,hue='IncomeGroup')  
plt.show()               # hue = colour for clear understanding
```

```python
vis7=sns.lmplot(data=df,x='InternetUsers',y='BirthRate',hue='IncomeGroup')  
plt.show()  
```
